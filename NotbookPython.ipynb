```python
#--------------------------
#VERIFICADOR DE DECOLAGEM
#--------------------------

#importação das bibliotecas
import random
from time import sleep

#crição das variáveis
tempint = random.randint(10, 35)
tempext = random.randint(-90, -40)
intestr = random.randint(0, 1)
energsist = random.randint(65, 100)
pretan = random.randint(75, 115)
modcri = 1

print('CALCULANDO DADOS...')
sleep(3)

#criação da tabela dos dados gerados
cabecalho = ['Parâmetro', 'Valor Calculado', 'Limite Seguro']
dados = [
    ['Temperatura Interna', tempint,  '15°C - 30°C' ],
    ['Temperatura Externa', tempext, '-80° até -50°'],
    ['Integridade Estrutural', intestr, 'Deve ser 1'],
    ['Energia do Sistema', energsist, '>= 70'],
    ['Pressão dos Tanques', pretan, '85 - 110 bar'],
    ['Módulo Crítico', modcri, 'Deve ser 1']
]

print(f'{cabecalho[0]:<25} {cabecalho[1]:<25} {cabecalho[2]:25}')
print('-' *70)

for linha in dados:
  print(f'{linha[0]:30} {linha[1]:<20} {linha[2]:<20}')

#autorização do foguete
if (tempint >= 15 and tempint <= 30) and (tempext >= -80 and tempext <= -50) and (intestr == 1) and (energsist >= 70) and (pretan >= 85 and pretan <= 110) and (modcri == 1):
  print('=' * 70)
  print('\n')
  print('O FOGUETE FOI AUTORIZADO PARA DECOLAGEM!!')
  print('\n')
  print('=' * 70)
else:
  print('=' * 70)
  print('\n')
  print('O FOGUETE NÃO FOI AUTORIZADO PARA DECOLAGEM!')
  print('\n')
  print('=' * 70)

#--------------------------
# ANÁLISE ASSISTIDA POR IA
#--------------------------

from sklearn.ensemble import IsolationForest

print("\nANÁLISE ASSISTIDA POR IA")

# criação de dados simulados para treinamento da IA
dados_treinamento = []

for i in range(100):

    t_int = random.randint(10,35)
    t_ext = random.randint(-90,-40)
    integ = random.randint(0,1)
    energia = random.randint(65,100)
    pressao = random.randint(75,115)
    modulo = 1

    dados_treinamento.append([
        t_int,
        t_ext,
        integ,
        energia,
        pressao,
        modulo
    ])

# treinamento do modelo
modelo = IsolationForest(contamination=0.05)
modelo.fit(dados_treinamento)

# dados do foguete gerados anteriormente
dados_foguete = [[
    tempint,
    tempext,
    intestr,
    energsist,
    pretan,
    modcri
]]

#--------------------------
# CLASSIFICAÇÃO DOS DADOS
#--------------------------

print("\nCLASSIFICAÇÃO DOS DADOS")

if energsist >= 70 and 85 <= pretan <= 110:
    print('Sistema Operacional Adequado!')
else:
    print('Sistema Operacional Inadequado!')


#--------------------------
# DETECÇÃO DE ANOMALIAS
#--------------------------

resultado = modelo.predict(dados_foguete)

print('\nDETECÇÃO DE ANOMALIAS')

if resultado[0] == -1:
    print('Anomalia Detectada!')
else:
    print('Nenhuma Anomalia Detectada')


#--------------------------
# SUGESTÕES DE RISCO
#--------------------------

print('\nSUGESTÕES DE RISCO')

if tempint > 30:
    print('Temperatura Interna Muito Elevada!')

if tempint < 15:
    print('Temperatura Interna Muito Baixa!')

if tempext < -80:
    print('Temperatura Externa Muito Baixa!')

if tempext > -50:
    print('Temperatura Externa Muito Elevada!')

if intestr != 1:
    print('Possível Falha Estrutural!')

if energsist < 70:
    print('Energia do Sistema Muito Baixa!')

if pretan < 85:
    print('Pressão dos Tanques Muito Baixa!')

if pretan > 110:
    print('Pressão dos Tanques Muito Alta!')

```
